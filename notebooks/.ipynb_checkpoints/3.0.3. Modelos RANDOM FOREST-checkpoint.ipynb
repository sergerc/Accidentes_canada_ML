---
Un [`random forest`](https://towardsdatascience.com/understanding-random-forest-58381e0602d2) consta de una gran cantidad de arboles de decisin individuales que operan en conjunto. Cada rbol individual en el bosque aleatorio escupe una prediccin de clase y la clase con ms votos se convierte en la prediccin de nuestro modelo. 




**METODOS DE CLASIFICACIN**
---

Como ya hemos comentado en notebooks anteriores. El principal problema que nos encontramos en este dataset es el desbalanceo de datos. La gran desproporcionalidad que existe entre los accidentes de trafico mortales de los que no son mortales es de 2% a 98%. Si bien, es cierto que hemos actuado frente a este desbalanceo mediante una `tcnica de muestreo aleatorio`en el notebook 
[2.0.3.Preprocesing.](http://localhost:8888/notebooks/Desktop/CUNEF/Practica%20Machine%20Learning/project_template/notebooks/2.0.3.%20Preprocesing.%20SPLIT%2C%20NA%2C%20OUTLIER.%20.ipynb) 

Sin embargo, nos seguimos encontrado con este problema en la muestra de test seleccionada. Esto afectar a nuestras mtricas, en nuestro caso, no representarn la precisin ni el recall adecuado del modelo.

Sabiendo el problema al que nos enfrentamos y teniendo como objetivo la prediccin de la mortalidad en un accidente de trfico. (Problema de clasificacin), enumeramos a continuacin las mtricas que vamos a utilizar: 

- `Precisin:` Se define como la divisin de los verdaderos positivos (TP) de entre todos los positivos predichos (TP + FP).En nuestro caso en particular, ser este dato el que este desequilibrado, ya que hay una desproporcionalidad de la muestra de muertes y no muertes. 

$$\frac{\textrm{TP}}{\textrm{TP+FP}}$$


- `Recall:` Se define como todos los ejemplos predichos que pertenecen a una clase (TP) de entre todos los positivos predichos (TP + FN). 

$$\frac{\textrm{TP}}{\textrm{TP+FN}}$$

- [`Accuracy`](https://machinelearningmastery.com/failure-of-accuracy-for-imbalanced-class-distributions/): Es una metrica que resume el rendimiento de un modelo de clasificacin con el nmero de predicciones correctas dividido por el nmero total de predicciones. Debido a que la prediccin no es una mtrica fiable, esta mtrica tampoco lo ser. 

$$\frac{\textrm{TP+TN}}{\textrm{TP+TN+FP+FN}}$$

- [`Matriz de confusin:`](https://medium.com/analytics-vidhya/accuracy-on-imbalanced-datasets-and-why-you-need-confusion-matrix-937613bf89bf) La mtrica mas simple y al mismo tiempo la ms efectiva para mirar el desempeo de los modelos en casos de dataset imbalanceados. Nos muestra la relaccin que existe entre positivos acertados (TP), positivos fallados (FN), negativos acertados(TN), y negativos fallados (FP). 


- [`F1-Score:`](https://www.iartificial.net/precision-recall-f1-accuracy-en-clasificacion/) La F-Score combina las dos mtricas de precisin y recall dentro de un mismo valor, aadiendo un parmetro `Beta`, que aplica un peso mayor a la precisin (Beta < 1) o un peso mayor al recall (Beta > 1). En nuestro caso, aplicaremos una F-1, ya que el peso lo aplicaremos en el modelo.

$$2 \times\frac{\textrm{Precision} \times \textrm{Recall}}{\textrm{Precision} + \textrm{Recall}}$$
    
   
- [`La curva ROC:`](https://machinelearningmastery.com/roc-curves-and-precision-recall-curves-for-imbalanced-classification/) Resume el rendimiento del modelo en la clase binaria positiva. Es una herramienta de diagnstico popular para clasificadores en problemas de prediccin binaria balanceados y desequilibrados por igual porque no est sesgada hacia la clase mayoritaria o minoritaria.


- [`Precision-Recall AUC:`](https://www.davidsbatista.net/blog/2018/08/19/NLP_Metrics/) Representa la precisin y el recall para diferentes umbrales de probabilidad. Es una tecnica muy efectiva para dataset desequilibrados debido a su enfoque en la clase minoritaria, es decir, en la mortalidad. 


- [`Cumulative Gain Curve`](https://towardsdatascience.com/meaningful-metrics-cumulative-gains-and-lyft-charts-7aac02fc5c14#:~:text=The%20cumulative%20gains%20curve%20is,target%20according%20to%20the%20model.) Evalua el rendimiento del modelo comparando los resultado con la seleccin aleatoria.


- [`Lift Curve`](https://towardsdatascience.com/meaningful-metrics-cumulative-gains-and-lyft-charts-7aac02fc5c14#:~:text=The%20cumulative%20gains%20curve%20is,target%20according%20to%20the%20model.) Mide la cantidad de ganancias que tiene nuestro modelo aplicandolo con respecto a la seleccin aleatorio.  Nos indica cuanto vale la pena implementar ese modelo.

#### Importo Librerias

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
import pyarrow.csv as pv
import pyarrow.parquet as pq
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import pyplot
from sklearn import preprocessing
import pandas as pd
import pyarrow.csv as pv
import pyarrow.parquet as pq
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc, \
                            silhouette_score, recall_score, precision_score, make_scorer, \
                            roc_auc_score, f1_score, precision_recall_curve
from sklearn.metrics import accuracy_score, roc_auc_score, \
                            classification_report, confusion_matrix
from sklearn.metrics import plot_confusion_matrix
from sklearn import tree
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.metrics import fbeta_score
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import GridSearchCV
import time
import pickle
import scikitplot as skplt
```

#### Importamos funciones: 

```python
def classify_metrics(modelo, x_test, y_test, predict): 
    
    '''Imprimimos el reporte de clasificacin:  '''
    
    print("Classification Report")
    print(classification_report(y_test, predict))
    
    '''Imprimimos el Accuracy: '''
    print("Accuracy:", metrics.accuracy_score(y_test, predict))
    
    '''Imprimimos el F1-score'''
    print("F1-Score:", metrics.f1_score(y_test, predict))
    
    print("---")
    
    '''Matriz de confusin: '''
    titles_options = [("Confusion matrix, without normalization", None),
                  ("Normalized confusion matrix", 'true')]
    for title, normalize in titles_options:
        disp = plot_confusion_matrix(modelo, x_test, y_test,
                                     display_labels=["mortal", "no mortal"],
                                     cmap=plt.cm.Blues,
                                     normalize=normalize)
        disp.ax_.set_title(title)

        print(title)
        print(disp.confusion_matrix)

    plt.show()
    
    yhat = modelo.predict_proba(x_test)
    pos_pred = yhat[:,1]
    
    # calculate roc curves
    fpr, tpr, thresholds = roc_curve(y_test, pos_pred)
    # plot the roc curve for the model
    plt.plot([0,1], [0,1], linestyle='--', label='No Skill')
    plt.plot(fpr, tpr, marker='.', label='Random Forest')
    # axis labels
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend()
    # show the plot
    plt.show()

    
    # calculate the g-mean for each threshold
    gmeans = np.sqrt(tpr * (1-fpr))
    
    # locate the index of the largest g-mean
    ix = np.argmax(gmeans)
    print('Best Threshold=%f, G-Mean=%.3f' % (thresholds[ix], gmeans[ix]))
    
    '''Dibujando la curva ROC: '''
    print("CURVA ROC con THRESHOLD:")
    # plot the roc curve for the model
    plt.plot([0,1], [0,1], linestyle='--', label='No Skill')
    plt.plot(fpr, tpr, marker='.', label='Modelo')
    plt.scatter(fpr[ix], tpr[ix], s=100, marker='o', color='black', label='Best')
    # axis labels
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend()
    # show the plot
    plt.show()
    
    '''Calculando el area AUC: '''
    print("AREA AUC: ")
    # calculate roc auc
    roc_auc = roc_auc_score(y_test, pos_pred)
    print('ROC AUC %.3f' % roc_auc)
    
    '''Calculamos la curva pr: '''
    # calculate pr-curve
    precision, recall, thresholds = precision_recall_curve(y_test, pos_pred)
    print("PRECISSION/RECALL CURVE: ")
    # plot the roc curve for the model
    no_skill = len(y_test[y_test==1]) / len(y_test)
    plt.plot([0,1], [no_skill,no_skill], linestyle='--', label='No Skill')
    plt.plot(recall, precision, marker='.', label='Random Forest')
    # axis labels
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.legend()
    # show the plot
    plt.show()
    
    '''Cumulative Gain Curve: '''
    skplt.metrics.plot_cumulative_gain(y_test, yhat)
    plt.show()
    
    '''Lift Curve:'''
    skplt.metrics.plot_lift_curve(y_test, yhat)
    plt.show()
```

#### Importamos el train y test seleccionado:

```python
train = pd.read_parquet('../data/trainselect1.parquet')
test = pd.read_parquet('../data/testselect1.parquet')
```

Divido entre 'xtrain' e 'ytrain'

```python
x_train = train.drop('c_sev',axis=1)
x_test = test.drop('c_sev',axis=1)
y_train = train['c_sev']
y_test = test['c_sev']
```

#### RANDOM FOREST :

Le introducimos el parametro class_weight, para evitar la desporporcionalidad del dataset, en este sentido, le aplicamos a 1 un peso elevado para ver como reacciona el modelo. Ya que, con el 40,60 que hemos empleado en la [`regresin lineal`](http://localhost:8888/notebooks/Desktop/CUNEF/Practica%20Machine%20Learning/project_template/notebooks/3.0.2.%20Modelos.%20Linear%20Regression.%20.ipynb) no nos sirve. Parece que el modelo no aprende de los accidentes. 

```python
%%time
rf = RandomForestClassifier(class_weight={0:1,1:1000})
start = time.time()
model3 = rf.fit(x_train, y_train)
end= time.time()
y_pred_rf = model3.predict(x_test)
print(y_pred_rf)
print("Tiempo de ejecucin fit: ", end-start)
```

#### Metricas: 

```python
classify_metrics(model3, x_test, y_test, y_pred_rf)
```

#### AJUSTAMOS EL MODELO CON EL THRESHOLD OBTENIDO

```python
'''Definimos los thresholds, el gmeans, el fpr y el tpr: '''
yhat = model3.predict_proba(x_test)
pos_pred = yhat[:,1]
fpr, tpr, thresholds = roc_curve(y_test, pos_pred)
gmeans = np.sqrt(tpr * (1-fpr))
ix = np.argmax(gmeans)
```

```python
'''Realizamos la prediccin aplicando el threshold: '''
y_pred_final = (pos_pred >= thresholds[ix]).astype(int)
y_pred_final
```

```python
x = metrics.confusion_matrix(y_test,y_pred_final, normalize='true')
sns.heatmap(x, annot=True, center=1)
```

#### Comentarios: 

Podemos ver como las predicciones del random forest en este caso son bastante pobres... NO nos sirve para determinar las muertes por encima de las no muertes. Con el parametro class_weight en un valor extremo (atipico) no aplicando nada de peso a los 1 y todo el peso a los 0. Solo podemos sacar una proporcin de fallos de 26 a 39 lo que nos daria un 13% menos de dinero activado para accidentes. 

POdemos ver como la curva PR en su primer tramo esta mas a favor del recall que de la precisin, aunque esta a medida que avanza va cambiando de tendencia. Aun asi no podemos aceptarla. 

La curva roc arroja un area un peln menor que la regresin lineal.

#### Queremos determinar la importancia de las variables sobre la variable objetivo: 

Parece que ha cambiado el peso de la variable con respecto al resto de los modelos. En el random forest tiene mas peso las horas, las semanas y los meses que la configuracin del accidente o la configuracin del asfalto, que pasa a tener una importancia media. 

```python
feature_importance = model3.feature_importances_
feature_importance = 100.0 * (feature_importance / feature_importance.max())
sorted_idx = np.argsort(feature_importance)
pos = np.arange(sorted_idx.shape[0]) + .5
# plt.subplot(1, 2, 2)
plt.figure(figsize=(15, 8))
plt.barh(pos, feature_importance[sorted_idx], align='center')
plt.yticks(pos, x_train.keys()[sorted_idx])
plt.xlabel('Relative Importance')
plt.title('Variable Importance')
plt.show()
```

#### RANDOM FOREST CON VALIDACIN CRUZADA. 

La [`validacin cruzada o cross-validation`](https://es.wikipedia.org/wiki/Validaci%C3%B3n_cruzada) es una tcnica utilizada para evaluar los resultados de un anlisis estadstico y garantizar que son independientes de la particin entre datos de entrenamiento y prueba. Consiste en repetir y calcular la media aritmtica obtenida de las medidas de evaluacin sobre diferentes particiones.

En este dataset nos volvemos a topar con el problema de desbalanceo del dataset sobre todo en la parte del test. Para ello, el cross validation tiene que adaptarse a esta desproporcin. Utilizaremos el 'RepeatedStratifiedKFold' que estratifica la muestra de train manteniendo la proporcin de muertes y no muertes. 

Por otro lado, combinaremos el cross validation con el GridSearch, para encontrar los mejores parametros que se adapten a nuestro modelo. En este sentido, comentamos los parametros sobre los que vamos a buscar su mejor optimizacin: 

- `n_estimators:` El numero de arboles que queremos en nuestro "bosque". Usaremos un range que recorra de 10 en 10 desde un punto de partida de 100 arboles hasta un punto final de 200. 


- `criterion:` La funcin que divide las ramas de los arboles, podemos escoger entre "gini" y "entropia". 


- `max_depth:` El numero maximo de particiones del arbol, creemos que lo justo sera un arbol entre 4 y 10 particiones. 


- `min_samples_split:` En numero minimo de elementos que tiene que tener una rama antes de partirse. Por defecto es dos, y como nuestro problema es binario. No pondremos nada. 


- `class_weight:` Le introducimos el parametro class_weight, para evitar la desporporcionalidad del dataset, en este sentido, le aplicamos a 1 un peso elevado para ver como reacciona el modelo. Ya que, con el 40,60 que hemos empleado en la regresin lineal no nos sirve. Parece que el modelo no aprende de los accidentes


Asimismo, utilizaremos el [`gridsearch`](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html) para encontrar los mejores parametros, esta funcin tambien tiene parametros que son dignos de mencin: 

- `verbose:` Pondremos =1 para que nos de el tiempo de ejecucin del codigo. 


- `n_jobs:` Pondremos =-1 para que la maquina utilice todos los procesadores de la cpu. 

---

```python
parameters = {
    'n_estimators': np.arange(150,170,10),
    'criterion': ('gini', 'entropy'),
    'max_depth': np.arange(4,8,1),
    'class_weight': ('balanced', 'balanced_subsample', None)
}

rfc = RandomForestClassifier(class_weight={0:1,1:1000})

'''Asignamos la validacin cruzada: '''
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=12345)

'''Asignamos el GridSearch: '''
start=time.time()
gs = GridSearchCV(rfc, parameters, verbose=1, n_jobs=-1, cv=cv, scoring='roc_auc')
model3cv = gs.fit(x_train, y_train)
end=time.time()
y_pred_rfc_cv = model3cv.predict(x_test)

'''imprimimos los parametros: '''
print(model3cv.best_params_)
print(model3cv.best_score_)
print("Tiempo de ejecucin: ", end-start)
```

#### Guardamos el modelo con [`Pickle`](https://machinelearningmastery.com/save-load-machine-learning-models-python-scikit-learn/)

```python
filename = '../models/model3cv.sav'
pickle.dump(model3cv, open(filename, 'wb'))
```
