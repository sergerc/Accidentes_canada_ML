3.0.8. INTERPRETABILIDAD.
---

La [`interpretabilidad`](https://christophm.github.io/interpretable-ml-book/interpretability.html) en Machine Learning es el grado de comprensión que tiene un ser humano sobre la decisión que toma un algoritmo. Es decir, la explicación o el razonamiento que le da a la conclusión final. Muchas veces, un algoritmo puede predecir con fiabilidad un problema planteado con soluciones no contempladas por el ser humano a la hora de introducir los datos. Es por ello que queremos interpretar los datos que hemos obtenido de nuestro mejor modelo, [`LightGBM`](http://localhost:8888/notebooks/Desktop/CUNEF/Practica%20Machine%20Learning/project_template/notebooks/3.0.5.%20Modelos%20LightGBM.ipynb). 




Para interpretar nuestro modelo utilizaremos una combinación de los modelos de interpretación de [`ELI5`](https://eli5.readthedocs.io/en/latest/) y [`LIME`](https://lime.readthedocs.io/en/latest/) llamada [`SHAP o (SHapley Additive exPlanations)`](https://shap.readthedocs.io/en/latest/index.html) que contiene un enfoque de teoria de juegos para explicar el resultado de cualquier modelo de aprendizaje automático. 


El objetivo de `SHAP` es explicar la predicción de una instancia x calculando la contribución de cada característica a la predicción.

#### Modelo LightGBM + SHAP:

Conocemos que [`LightGBM`](https://lightgbm.readthedocs.io/en/latest/) tiene un parámetro que se puede utilizar de forma conjunta con SHAP. En la predicción de la probabilidad, podemos introducit el parámetro `pred_contrib = True` que te muestra los valores Shap para cada una de las observaciones. Lo que supondrá un mejor entendimiento para las personas de negocio.

Cargamos el modelo con pickle. 

Nos encontramos con un problema y es que, el bucle tarda mucho por lo que, únicamente usaremos esto para los 500 primeros casos. 

```python
%%time
filename = '../models/model5def.sav'
model5 = pickle.load(open(filename, 'rb'))
y_pred_lgbm = model5.predict(x_test)
print(y_pred_lgbm)
```

```python
%%time
# Escojo solo las 500 primeras muestras porque sino lleva mucho tiempo
pd_shap = pd.DataFrame(model5.predict(x_test.head(500), pred_contrib=True),
                       columns=list(x_test.columns)+['other']).drop(['other'], axis=1)
y_pred_lgbm_500 = pd.DataFrame(model5.predict(x_test.head(500)))
```

```python
%%time
pd_contrib = pd.DataFrame({'contrib_1_max1':['0'],'contrib_1_max2':['0'],'contrib_1_max3':['0'],
                          'contrib_0_max1':['0'],'contrib_0_max2':['0'],'contrib_0_max3':['0']})

for i in range (pd_shap.shape[0]):
    list_contrib_1 = list(pd_shap.iloc[i,:].sort_values(axis=0, ascending=False).index[-3:])
    list_contrib_0 = list(pd_shap.iloc[i,:].sort_values(axis=0, ascending=False).index[:3])
    pd_row = pd.DataFrame([list_contrib_1 + list_contrib_0])
    pd_row.columns = pd_contrib.columns
    pd_contrib = pd.concat([pd_contrib, pd_row], axis=0)

pd_contrib = pd_contrib.iloc[1:,].set_index(y_pred_lgbm_500.index)
pd_contrib = pd.concat([y_pred_lgbm_500, pd_contrib], axis=1)
```

Como podemos observar, en función de cada casuística las variables principales van cambiando. Si que podemos observar por ejemplo, que para los accidentes mortales `(0)`, es muy importante el tramo de la carretera `c_raln` o  el tipo de superficie `c_rsur` ademas del tipo de colisión `c_conf` que se ha tenido. 

Sin embargo, este análisis sirve para cada caso en particular, nosotros nos queremos enfocar en el global de la predicción.


A continuación cargamos JS que es el codigo de visualización para notebook, y mostramos diferentes gráficos con Shap para interpretar nuestro modelo. Usaremos los siguientes gráficos: 

- [`TREE SHAP`](https://medium.com/analytics-vidhya/shap-part-3-tree-shap-3af9bcd7cd9b): ·Es un algoritmo para calcular valores SHAP exactos para modelos basados en árboles de decisión.

---

```python
%%time
shap.initjs()
explainer = shap.TreeExplainer(model5)
shap_values = explainer.shap_values(x_test)
shap.summary_plot(shap_values[1], x_test)
```

```python
%%time
shap.initjs()
shap.summary_plot(shap_values, x_test)
```

#### COMENTARIOS: 

Esta gráfica nos muestra el valor del SHAP en el `eje de las X` mientras que en el `eje de las Y` nos muestra la importancia de las variables. Interpretariamos este gráfico de la siguiente manera: 

- La configuración del accidente `c_conf`es lo que mas contribuye a decidir si un accidente ha sido mortal o no. 


- El numero de personas implicadas en un accidente `c_nums` así como el numero de vehicuos `c_vehs` y el dia de la semana `c_wday` es lo que mas contribuye a que los valores del score sean cercanos a `0.Mortalidad`.


- Existen otras variables que contribuyen a que haya mortalidad en un accidente como es la señalización de tráfico `c_traf` o la zona de la carretera donde se produce el accidente de tráfico `c_rcfg`. 


A continuación escogemos las dos observaciones aleatorios predichas y vemos la contribución de cada varible al score obtenido. 


```python
shap.force_plot(explainer.expected_value[1], shap_values[1][281,:], x_test.iloc[281,:], link='logit')
```

```python
shap.force_plot(explainer.expected_value[1], shap_values[1][182,:], x_test.iloc[182,:], link='logit')
```

Hemos escogido la muestra 282 y la muetsra 182. Nos enseñan dos tipos de predicciones, la primera, muy a favor de la mortalidad, en la que los pesos mas importantes se encuentran en la configuración del accidente, y la superficie y la señalización de tráfico en menor medida. 

Por otro lado, el segundo gráfico, muy a favor de la no mortalidad, en la que tambien ha tenido que ver la configuración del accidente y la señalización del trafico. Por el contrario la superficie de la carretera contribuye a que el accidente sea mortal. 

Podemos componer este grafico de las primera 500 muestras, tal y como hemos hecho anteriormente en el LightGBM + SHAP: 

```python
%%time
shap.initjs()
shap.force_plot(explainer.expected_value[1], shap_values[1][:500,:], x_test.iloc[:500,:], link='logit')
```

El objetivo de la agrupación es encontrar grupos de instancias similares. Normalmente, la agrupación se basa en características. Las características son a menudo en diferentes escalas. Cada posición en el eje x es una instancia de los datos. Los valores SHAP azules aumentan la predicción, los valores rojos la disminuyen. Por lo que podremos decir, que los datos del gráfico situados a la derecha son mas propensos a tener accidentes mortales.
