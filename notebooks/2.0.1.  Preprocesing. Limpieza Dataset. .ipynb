#### 0. Carga Dataset

```python
df_acc = pd.read_parquet('../data/NCDB_1999_to_2014.parquet')
df_acc.columns = df_acc.columns.map(lambda x: x[:].lower())
df_acc
```

```python
df_acc.shape
```

#### 1. Eliminar duplicados.

Queremos saber qué filas están duplicadas y eliminarlas antes de agrupar algo que no deberiamos: 

```python
df_acc[df_acc.duplicated(keep="first")]
```

```python
df_acc_t = df_acc.drop_duplicates()
```

```python
df_acc_t.shape
```

---

#### 2.Eliminación de los valores anomalos. 

Mientras cargabamos y analizabamos el dataset en el notebook 01. Nos dimos cuenta de que hay valores anomalos que necesitamos eliminar. Se tratan de los coches del 2015 que tuvieron accidentes en 2014. Localizamos estas filas: 

```python
valores_anomalos = df_acc_t[(df_acc_t["v_year"] == "2015")&(df_acc_t["c_year"] == 2014)].index
valores_anomalos.shape
```

```python
df_acc_t = df_acc_t.drop(valores_anomalos)
```

```python
df_acc_t.shape
```

----

#### 3. Agrupación de los valores por accidente.

Tal y como hemos comentado en la carga de datos y pequeña exploración inicial (0.1.LOAD). Hay varias filas para cada accidente, por lo que, de cara a simplificar el dataset y poder quitar ruido, vamos a agrupar por las variables que comienzan por "c_".

```python
c_columns = ["c_year", "c_mnth", "c_wday", "c_hour", "c_sev",
            "c_vehs", "c_conf", "c_rcfg", "c_wthr", "c_rsur", 
             "c_raln", "c_traf"]
df_acc_group = df_acc_t.groupby(c_columns).count()["v_id"]
df_acc_group = df_acc_group.reset_index()
df_acc_group = df_acc_group.rename(columns={"v_id":"c_nums"})
```

```python
df_acc_group.shape
```

---

#### Guardamos el dataset 

```python
df_acc_group.to_parquet("../data/df_acc_1.parquet")
```
